```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#splitting data:training and testing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# DATA COLLECTION

```python
#loading
credit_data=pd.read_csv('/content/creditcard.csv')
```

# DATA PRE PROCESSING


```python
#check for missing values
credit_data.isnull().sum()
```

we can infer that the class attribute with 0 and 1 which implies 0 for normal transcations and 1 for fraud transcations

```python
#distribution of transcations
credit_data['Class'].value_counts()
```

This dataset is highly unbalanced cuz more than 99% data are in one particular class and if we fed this data into the machine and train it which results in  the machine mostly recognising the new data as a normal transcations.
Therefore it requires some processing.

```python
#separating the data for analysis
valid=credit_data[credit_data.Class==0]
not_valid= credit_data[credit_data.Class==1]
```

```python
print(valid.shape)
print(not_valid.shape)
```

```python
valid['Amount'].describe()
```

```python
not_valid['Amount'].describe()
```

insight- mean value for not_valid is greater than valid

```python
#comparing
credit_data.groupby('Class').mean()
```

sampling

```python
#building sample dataset with similar valid and not valid transactions
```

```python
#to create an balanced dataset
valid_sample=valid.sample(n=492)
```

```python
new=pd.concat([valid_sample,not_valid],axis=0)
```

```python
new.head()
```

```python
new.tail()
```

```python
new['Class'].value_counts()
```

```python
new.groupby('Class').mean()
```

compare it with the previously driven mean, if the mean values are of wider difference gap we have created a bad sample or else we can consider it as a good sample

# FEATURES AND TARGETS

```python
x=new .drop(columns='Class',axis=1)
y=new['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
#splitting data
#80% to train data and remaining for testing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

# MODEL TRANING

```python
model=LogisticRegression()
```

```python
#training
model.fit(x_train,y_train)
```

# MODEL EVALUATION

```python
#accuracy score for training
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
print("Accuracy of traing data: ",training_data_accuracy)
```

```python
#accuracy for test
x_test_prediction=model.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_prediction,y_test)
print("Accuracy of test data: ",testing_data_accuracy)
```

accuracy data is consistent so our system has trained well.
